```python
x = input("Введите число от 0 до 99:")  #создаем переменную, в которую пользователь должен ввести число
```

```python
#создаем 2 листа, первый для чисел от нуля до 20, так как их нельзя составить по одной схеме, второй для десятков
number = ["ноль","один","два","три","четыре","пять","шесть","семь","восемь","десять","одиннадцать","двенадцать","тринадцать","четырнадцать","пятнадцать","шестнадцать","семнадцать","восемнадцать","девятнадцать","двадцать"]
ten = ["двадцать","тридцать","сорок","пятьдесят","шестьдесят","семьдесят","восемьдесят","девяносто"]
```

```python
try:
 # используем try для того, чтобы исключить ошибку ValueError  
    if 0>int(x) or int(x)>99:
        print("Ваше число не входит в интервал!")
    elif 0<int(x)<=20:  #если число меньше либо равно 20, то просто выводим из листа соответствующее ему значение
        print(number[int(x)])
    elif 20<int(x)<=99:
        x_1 = str(x)  #создаем переменную в котором х превращается в строку(для того чтобы в последующем можно было вытащить первую и вторую цифру)
        print(ten[int(x_1[0]) -2], number[int(x_1[1])])  #сначала выводим десятки(вычитаем 2, так как начинаем с 20 и потому что в питоне считаем с 0
except ValueError:   #в случае если нам выдает ошибку ValueError программа продолжает с этой строки
    print("Вам нужно ввести число.")  
```
